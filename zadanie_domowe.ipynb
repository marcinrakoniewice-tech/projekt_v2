# Zadanie domowe

Dla wybranego projektu który realizujesz w ramach kursu (pomysł własny lub zainspirowany przez nas)

1. Dodaj Trello Board
1. "Otaskuj projekt" - wciel się w rolę "Product Owner" i utwórz kilka pierwszych zadań, które Twoim zdaniem należy wykonać w celu realizacji projektu
    - Task powinien mieć tytuł i krótki opis
1. Zaproś mentora (lub mentorów) do swojego Trello Boardu
    - Maciej - maciej@imprv.ai
    - Andrzej - andrzej@imprv.ai
1. Utwórz repozytorium dla tego projektu na GitHub
    - W zależności od preferencji może być to repozytorium prywatne lub publiczne
1. Zaproś do kontrybucji w repozytorium mentora (lub mentorów)
    - Maciej - maciej@imprv.ai
    - Andrzej - andrzej@imprv.ai
1. Wystaw Pull Request
    1. Utwórz `branch` odchodzący od gałęzi głównej (od `main`)
    1. Dokonaj zmian. Mogą to być zmiany w kodzie, notebookach lub jeśli nie masz na razie pomysłu
        to opisz krótko projekt w pliku `README.md`
    1. Wypchnij zmiany do repozytorium zdalnego
    1. Utwórz Pull Request 

1. Link do Trello Board i link do Pull Request'a wyślij w wiadomości na kanał `modul-11-zad-1`
